Aggregate function (sum,count,avg,min,max)

```sql
SELECT sum(inflation_adjusted_gross) from movie_gross
```

```sql
SELECT sum(inflation_adjusted_gross) as [total gross] , COUNT(*) as [ number of movies] from movie_gross
```

```sql
SELECT sum(inflation_adjusted_gross) as [total_gross],
    count(*) as [nuber of movies],
    avg(inflation_adjusted_gross) as [Average gross],
    min(inflation_adjusted_gross) as [Min Gross],
    max(inflation_adjusted_gross) as [Max Gross]
    from movie_gross
    
```

```sql
SELECT genre,DATEPART(year,release_date) as year,count(*) as [Number of Movies],sum(inflation_adjusted_gross) as [Total Gross by Genre] from movie_gross
group by genre,DATEPART(year,release_date) /* sum  genre  genre  group by */
order by [Total Gross by Genre] DESC
```

```sql
SELECT genre,DATEPART(year,release_date) as year,count(*) as [Number of Movies],sum(inflation_adjusted_gross) as [Total Gross by Genre] from movie_gross
group by genre,DATEPART(year,release_date) /* sum  genre  genre  group by */
order by genre, datepart(year, release_date) 
```

```sql
SELECT genre,DATEPART(year,release_date) as year,mpaa_rating,count(*) as [Number of Movies],sum(inflation_adjusted_gross) as [Total Gross by Genre] from movie_gross
group by genre,DATEPART(year,release_date),mpaa_rating 
order by genre, datepart(year, release_date) 
```

```sql
SELECT genre,count(*) as no_movies,sum(inflation_adjusted_gross) as [SUM GROSS],avg(inflation_adjusted_gross) as [AVG GROSS]from movie_gross
    GROUP by genre
    having count(*) > 100
    ORDER by SUM(inflation_adjusted_gross) DESC
```

```sql
SELECT COUNT(genre) as [Comedy and Action] from movie_gross WHERE genre IN ('Comedy','Action')
```
