```sql
 SELECT genre,DATEPART(YEAR,release_date) YEAR , SUM(inflation_adjusted_gross) gross
    from movie_gross WHERE genre in ('Action','Adventure') and year(release_date) > 2010
    GROUP BY genre,DATEPART(YEAR,release_date)
```

```sql
with ac as (
    SELECT genre,DATEPART(YEAR,release_date) YEAR , SUM(inflation_adjusted_gross) gross
    from movie_gross WHERE genre in ('Action') and year(release_date) > 2010
    GROUP BY genre,DATEPART(YEAR,release_date)
),
ad as (

SELECT genre,YEAR(release_date) YEAR , SUM(inflation_adjusted_gross) gross
from movie_gross WHERE genre in ('Adventure') and YEAR(release_date) > 2010
GROUP by genre, YEAR(release_date))

SELECT ac.YEAR , ac.gross [Action] ,ad.gross [Adventure]
from ac inner JOIN ad on ac.YEAR = ad.[YEAR]
```

```sql
SELECT movie_title,total_gross,(select sum(total_gross) from movie_gross where YEAR(release_date) = 2016) as [Sum total gross],
total_gross *1.0 / (select sum(total_gross) from movie_gross where YEAR(release_date) = 2016)*100
 as [% from total gross] 
 from movie_gross WHERE YEAR(release_date) = 2016
```

```sql
SELECT movie_title,total_gross,(select avg(total_gross) from movie_gross where YEAR(release_date) = 2016) as [Avg of Total Gross],YEAR(release_date)
from movie_gross WHERE (total_gross > (select avg(total_gross) from movie_gross where YEAR(release_date) = 2016)) and YEAR(release_date) = 2016
```

```sql
select movie_title,release_date,inflation_adjusted_gross,
RANK() OVER (PARTITION BY YEAR(release_date) order by inflation_adjusted_gross desc) gross_rank
 from movie_gross
 WHERE YEAR(release_date) > 2010
```

```sql
WITH cte as (
select movie_title,release_date,inflation_adjusted_gross,
RANK() OVER (PARTITION BY YEAR(release_date) order by inflation_adjusted_gross desc) gross_rank
 from movie_gross
 WHERE YEAR(release_date) > 2010)
 SELECT * FROM cte where gross_rank <=3
```

```sql
SELECT director,COUNT(*) no_movies from director GROUP by director ORDER by no_movies DESC
```

```sql
with cte as (
    SELECT director,COUNT(*) no_movies from director GROUP by director 
)
select *,
    RANK() over (order by no_movies desc) rank,
    DENSE_RANK() over (order by no_movies desc) rank1
from cte
```

```sql
SELECT movie_title,release_date,inflation_adjusted_gross,
NTILE(4) OVER(order by inflation_adjusted_gross desc) as quartile
from movie_gross
```

```sql
with cte as (
    SELECT movie_title,release_date,inflation_adjusted_gross,
NTILE(4) OVER(order by inflation_adjusted_gross asc) as quartile
from movie_gross
)
select * from cte WHERE quartile = 1 and inflation_adjusted_gross <> 0
```
